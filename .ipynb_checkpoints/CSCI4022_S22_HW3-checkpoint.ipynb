### A)
Write a function `kmeans` that takes in a data frame, and uses a predefined distance to run k-means until convergence.  

- Check for convergence by calculating the reconstruction error after each step, and stop when the reconstruction error has changed by no more than `tol` from the previous step.
- Choose $k$ rows form the data frame at random to be the initial centroids of the clusters

Use the in-class notebook code slide, and/or textbook to guide you.

```python
def distance(x1, x2, x3, y1, y2, y3):
    return np.sqrt((x1 - y1) ** 2 + (x2 - y2) ** 2 + (x3 - y3) ** 2)

def kmeans(df, k, tol = 0.05):
    tempDifference = 0
    reconError = 100
    differences = [0] * k
    minDifference = 10000
    tempIndex = 0
    totalDifferences = []
    centroids = [0] * k
    clusters = [-1] * len(df)
    randomNums = rd.sample(range(0, len(df)), k)
    for i in range (k):
        centroids[i] = df.iloc[randomNums[i]]
    for index in range (len(df)):
        for j in range (k):
            tempDifference =  distance(df.iloc[index]['HP'], df.iloc[index]['Attack'], df.iloc[index]['Defense'], centroids[j]['HP'], centroids[j]['Attack'], centroids[j]['Defense'])
            differences[j] = tempDifference
        minValue = min(differences)    
        totalDifferences.append(minValue)
        minIndex = differences.index(minValue)
        clusters[index] = minIndex   
  
    oldError = sum(totalDifferences) / len(totalDifferences)    
    while(reconError > tol):
        counter = -1
        newCentroids = [(0,0,0)] * k
        tempDifferences = 0
        differences = [0] * k
        minDifference = 10000
        differencesForClusters = []
        tempIndex = 0 
        for a in range (k):
            counter += 1
            averageHP = []
            averageAt = []
            averageDe = []
            for b in range (len(df)): 
                if(clusters[b] == a):
                    averageHP.append((df.iloc[b]['HP']))
                    averageAt.append((df.iloc[b]['Attack']))
                    averageDe.append((df.iloc[b]['Defense']))     
            newCentroids[a] = ((sum(averageHP) / len(averageHP)), (sum(averageAt) / len(averageAt)), (sum(averageDe) / len(averageDe)))  
        for index in range (len(df)):
            for j in range (k):
                tempDifference =  distance(df.iloc[index]['HP'], df.iloc[index]['Attack'], df.iloc[index]['Defense'], newCentroids[j][0], newCentroids[j][1], newCentroids[j][2])
                differences[j] = tempDifference
            minValue = min(differences)    
            differencesForClusters.append(minValue)
            minIndex = differences.index(minValue)
            clusters[index] = minIndex 
        newError = sum(differencesForClusters) / len(differencesForClusters)
        reconError = abs(oldError - newError)
        if(reconError < tol):
            return newCentroids, clusters, newError
        oldError = newError
```

# Problem 3 (K-Means and Distance Measures; 18 pts)

If you don't know much about Pokémon, you should know the following:
- Mudkip is the cutest
- The "type" column matters quite a bit

... and we didn't use the "type" column!  To do so, we can modify k-means, but it requires us also define a *distance* that we can input types into.  A distance that can be calculated using categories?  Sounds like Jaccard similarity!  Suppose we decide to include the Jaccard distance (recall: it's $1-sim$) in addition to Euclidean distance.  The distance between two Pokémon is some combination of the distance between their numeric stats and the distances between their typings, just as in problem **1**!

```python
def euc_distance(pokemon1, pokemon2):
    distance = np.sqrt(np.sum( ((pokemon1['HP'] - pokemon2['HP']) ** 2) + ((pokemon1['Attack'] - pokemon2['Attack']) ** 2) + ((pokemon1['Defense'] - pokemon2['Defense']) ** 2)) )
    return distance
def jacc_distance(pokemon1, centroid):
    types = []
    types.append(pokemon1['Type 1'])
    types.append(pokemon1['Type 2'])
    for x in centroid:
        types.append(x)
    filteredTypes = [pokeType for pokeType in types if (pokeType != "nan")]
    filteredTypes = set(filteredTypes)
    numer = 0
    if pokemon1['Type 1'] in filteredTypes:
        numer += 1
    if pokemon1['Type 2'] in filteredTypes:
        numer += 1
    jaccard = numer / len(filteredTypes)
    return (1 - jaccard)
```

Using $k=4$ and at least 10 runs, cluster the Pokémon data set where the distance between two Pokémon is the *sum* of their Euclidean ($L_2$) and Jaccard distances.  Represent clusters by their centroids for numeric stats, and by the union of all of their typings for type.  Make a scatter plot of the resulting assignments on an (Attack, Defense) axis.

Does it look like typing is mattering?

```python
def kmeansJacc(df, k, tol = 0.05):
    tempDifference = 0
    tempJacDistance = 0
    reconError = 100
    differences = [0] * k
    differencesJac =[0] * k
    minDifference = 10000
    tempIndex = 0
    totalDifferences = []
    centroids = [0] * k
    centroidTypes = []
    clusters = [-1] * len(df)
    randomNums = rd.sample(range(0, len(df)), k)
    for i in range (k):
        centroids[i] = df.iloc[randomNums[i]]
        centroidTypes.append(df.iloc[randomNums[i]]["Type 1"])
        centroidTypes.append(df.iloc[randomNums[i]]["Type 2"])
    for index in range (len(df)):
        for j in range (k):
            tempDifference = euc_distance(df.iloc[index], centroids[j])
            tempJacDistance = jacc_distance(df.iloc[index], centroidTypes)
            differences[j] = tempDifference
            differencesJac[j] = tempJacDistance
        minValue = min(differences)    
        minJacc = min(differencesJac)
        totalDifferences.append(minValue + minJacc)
        minIndex = differences.index(minValue)
        clusters[index] = minIndex  
    oldError = sum(totalDifferences) / len(totalDifferences)    
    
    while(reconError > tol):
        counter = -1
        newCentroids = [(0,0,0)] * k
        newCentroidTypes = []
        tempDifferences = 0
        differences = [0] * k
        minDifference = 10000
        differencesForClusters = []
        tempIndex = 0 
        for a in range (k):
            counter += 1
            averageHP = []
            averageAt = []
            averageDe = []
            for b in range (len(df)): 
                if(clusters[b] == a):
                    averageHP.append((df.iloc[b]['HP']))
                    averageAt.append((df.iloc[b]['Attack']))
                    averageDe.append((df.iloc[b]['Defense']))  
                    newCentroidTypes.append((df.iloc[b]['Type 1']))
                    newCentroidTypes.append((df.iloc[b]['Type 2']))
            newCentroids[a] = ((sum(averageHP) / len(averageHP)), (sum(averageAt) / len(averageAt)), (sum(averageDe) / len(averageDe)))  
        for index in range (len(df)):
            for j in range (k):
                tempDifference = euc_distance(df.iloc[index], centroids[j])
                tempJacDistance = jacc_distance(df.iloc[index], newCentroidTypes)
                differences[j] = tempDifference
            minValue = min(differences)    
            minJacc = min(differencesJac)
            totalDifferences.append(minValue + minJacc)
            minIndex = differences.index(minValue)
            clusters[index] = minIndex  
        newError = sum(totalDifferences) / len(totalDifferences)
        reconError = abs(oldError - newError)
        if(reconError < tol):
            return newCentroids, clusters, newError
        oldError = newError
```

### B) 

Your answer in part **3A** should be "certainly not!"  Since Jaccard distances all live in $[0,1]$ and each numerical statistic can range from $[1,256]$, the distances between those stats will totally dominate Jaccard.   But that's no problem!  We can either rescale the *data*, where we shrink stats down to a $[0,1]$ or similar scale, or we can rescale our *distance* to weight some features/columns more than others.  

Suppose we decide we want to "weight" Jaccard distance $d_J$ as 90% of the calculation and keep the Euclidean stats-distance $d_E$ between Pokémon as the remaining 10%.  We could define:

$$d(x,y)=.9 d_J(x,y) + .1 d_E(x,y)$$

as our distance!

```python
def combined_dist(pokemon1, pokemon2, a): #a is the "percentage" that we'll weight Jaccard
    dist=100*(1-a)*euc_distance(pokemon1, pokemon2)+100*a*jacc_distance(pokemon1, pokemon2)
    return dist


def kmeansJaccNew(df, k, weight, tol = 0.05):
    tempDifference = 0
    tempJacDistance = 0
    reconError = 100
    differences = [0] * k
    differencesJac =[0] * k
    minDifference = 10000
    tempIndex = 0
    totalDifferences = []
    centroids = [0] * k
    centroidTypes = []
    clusters = [-1] * len(df)
    randomNums = rd.sample(range(0, len(df)), k)
    for i in range (k):
        centroids[i] = df.iloc[randomNums[i]]
        centroidTypes.append(df.iloc[randomNums[i]]["Type 1"])
        centroidTypes.append(df.iloc[randomNums[i]]["Type 2"])
    for index in range (len(df)):
        for j in range (k):
            tempDifference = euc_distance(df.iloc[index], centroids[j])
            tempJacDistance = jacc_distance(df.iloc[index], centroidTypes)
            differences[j] = tempDifference
            differencesJac[j] = tempJacDistance
        minValue = min(differences)    
        minJacc = min(differencesJac)
        totalDifferences.append((100 * (1 - weight) * minValue) + (100 * weight * minJacc))
        minIndex = differences.index(minValue)
        clusters[index] = minIndex  
    oldError = sum(totalDifferences) / len(totalDifferences)    
    
    while(reconError > tol):
        counter = -1
        newCentroids = [(0,0,0)] * k
        newCentroidTypes = []
        tempDifferences = 0
        differences = [0] * k
        minDifference = 10000
        differencesForClusters = []
        tempIndex = 0 
        for a in range (k):
            counter += 1
            averageHP = []
            averageAt = []
            averageDe = []
            for b in range (len(df)): 
                if(clusters[b] == a):
                    averageHP.append((df.iloc[b]['HP']))
                    averageAt.append((df.iloc[b]['Attack']))
                    averageDe.append((df.iloc[b]['Defense']))  
                    newCentroidTypes.append((df.iloc[b]['Type 1']))
                    newCentroidTypes.append((df.iloc[b]['Type 2']))
            newCentroids[a] = ((sum(averageHP) / len(averageHP)), (sum(averageAt) / len(averageAt)), (sum(averageDe) / len(averageDe)))  
        for index in range (len(df)):
            for j in range (k):
                tempDifference = euc_distance(df.iloc[index], centroids[j])
                tempJacDistance = jacc_distance(df.iloc[index], newCentroidTypes)
                differences[j] = tempDifference
            minValue = min(differences)    
            minJacc = min(differencesJac)
            totalDifferences.append(((100 * (1 - weight)) * minValue) + (100 * weight * minJacc))
            minIndex = differences.index(minValue)
            clusters[index] = minIndex  
        newError = sum(totalDifferences) / len(totalDifferences)
        reconError = abs(oldError - newError)
        if(reconError < tol):
            return newCentroids, clusters, newError
        oldError = newError
```

Finally, using $k=4$ and at least 10 runs, cluster the Pokémon data set where the Jaccard distance between Pokemon is given a 99.5% of the weight of the distance function.

Make a scatter plot of the resulting assignments on an (Attack, Defense) axis.

```python
centroidsList = []
clustersList = []
meanerrorList = []
for i in range(10):
    centroids, clusters, meanerror = kmeansJaccNew(df, 4, .995, 0.05)
    centroidsList.append(centroids)
    clustersList.append(clusters)
    meanerrorList.append(meanerror)
    
minMeanerror = min(meanerrorList)    
minIndex = meanerrorList.index(minMeanerror)

centroids = centroidsList[minIndex]
clusters = clustersList[minIndex]
meanerror = meanerrorList[minIndex]

plt.scatter([x for x in df['Attack']], [x for x in df['Defense']], c = clusters)
plt.scatter([x[1] for x in centroids], [x[2] for x in centroids], c = 'red', marker='+')
```
