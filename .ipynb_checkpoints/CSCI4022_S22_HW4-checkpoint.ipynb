**Part A**:  Write (by hand) a function `GMM(dat, k, init=...)` that performs the EM algorithm for a Gaussian mixture model on a given set of $(X,Y)$ locational data.  You may include random initialization inside the function or as an input.

You should run the algorithm for sufficient steps until convergence, and should return at a minimum, the means, 2x2 covariance matrices, and mixture probabilities of each component.  You may also want to return the responsibilities of each datum to the k components for later use, but this is less necessary.

```python
import copy
import math

def GMM(df, k, tol):

    randomPis = [0]
    randomNums = []
    p_class_given_data = []
    p_data_given_class = []
    randomVars = []
    distances = [0] * len(df)
    avgDistances = [0] * k
    means = []
    for i in range(k):
        randomLat = rd.uniform(df['lat'].min(), df['lat'].max())
        randomLng = rd.uniform(df['lng'].min(), df['lng'].max())
        means.append([randomLat, randomLng])
        randomVars.append([[1, 0], [0, 1]])
        randomNums.append(np.random.randint(1, 100))

    randomPis = [num / sum(randomNums) for num in randomNums]
  
    for i in range(k):
        for j in range(len(df)):
            distances[j] = np.sqrt( ( ( (df.loc[j]['lat'] - means[i][0]) ** 2) + (df.loc[j]['lng'] - means[i][1]) ** 2) * randomPis[i])
        avgDistances[i] = sum(distances) / len(distances)
    oldMeanDif = sum(avgDistances) / k   
    
    while(1): 
        for j in range(k):
            p_data_given_class.append(stats.multivariate_normal.pdf(x = (df), mean = means[j], cov = np.sqrt(randomVars[j]) ))
            p_class_given_data.append(p_data_given_class[j]) 

        sumOfPData = sum(p_class_given_data)
        p_class_given_data = [x / sumOfPData for x in p_class_given_data]
        n_class = []
        
        for a in range(k):
            n_class.append(np.sum(p_class_given_data[a]))
            randomPis[a] = n_class[a] / len(df)

        for b in range(k):       
            means[b] = ( (1 / n_class[b]) * np.sum(p_class_given_data[b] * df['lat']), (1 / n_class[b]) * np.sum(p_class_given_data[b] * df['lng']) )

            randomVars[b][0][0] = np.var(df['lat'], ddof = 1)
            randomVars[b][1][1] = np.var(df['lng'], ddof = 1)
            diag = abs(np.sum( ( (df['lat'] - means[b][0]) * (df['lng'] - means[b][1]) ) / (len(df) - 1) ))
            randomVars[b][1][0] = diag
            randomVars[b][0][1] = diag
            
        for i in range(k):
            for j in range(len(df)):
                distances[j] = np.sqrt( ( ( (df.loc[j]['lat'] - means[i][0]) ** 2) + (df.loc[j]['lng'] - means[i][1]) ** 2) * randomPis[i])
            avgDistances[i] = sum(distances) / len(distances)
            
        newMeanDif = sum(avgDistances) / k
        meanDif = abs(newMeanDif - oldMeanDif)
        if(meanDif < tol):
            return p_class_given_data, means, randomVars, randomPis
        else:
            oldMeanDif = newMeanDif
```

**Part B**: 

Run your algorithm for $k=[1,2,3,4,5,6,7,8]$, then create an **elbow plot** of the weighted-mean-distance to component for each choice of $k$.

To compute this, take a clustering and its responsibilities $p_{mi}$, then compute the average distance between each point and each cluster, weighted by $p_{mi}$:

$$ \bar{D} = \sqrt{\frac{\sum^n_{i=1}\sum^k_{m=1} p_{mi} (x_i-\mu_m)^2}{n}}$$

(**Idea**: this averages the distance between each point and the 'closest' cluster, but allows for the soft-clustering nature of points to compute distances between multiple clusters.)


Create your elbow plot by making a line plot of the values of $\bar{D}$ against $k$.

```python
avgDistances = []

for i in range(1,9):
    data, means, var, pis = GMM(df, i, 0.01)
    distances = [0] * len(df) * (i)
    for a in range(i):
        for j in range(len(df)):
            distances[(a * len(df)) + j] = np.sqrt( ( ((df.loc[j]['lat'] - means[a - 1][0]) ** 2) + ((df.loc[j]['lng'] - means[a - 1][1])) ** 2) * pis[a])
    avgDistances.append(sum(distances) / len(distances)) 
    
plt.plot((range(1, 9)), avgDistances)    
```
